## Sample dataset

- menyiapkan sebuah variabel yang menampung beberapa nilai 

- menggunakan fungsi np.linspace

- np.meshgrid menghasilkan 2 buah array

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wireframe

- hasil nya bisa di rotasikan

- dikarenakan kita menggunakan matplotlib notebook

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

## 3D Surface Plot

- sama dengan sebelumnya, hanya saja ini membuat graph menjadi 2D

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface plot')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface plot')

plt.show()
```
