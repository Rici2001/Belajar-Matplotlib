## Simple Line Plot

-Menyiapkan sample data set ynag kita generate dengan menggunakan fungsi np.arange yang menyertakan 3 buat parameter

```python
x = np.arange (0.0, 2.0, 0.01)
x
```

-Kemudian kita akan membuat data set yang kedua yang kita tampung di variable 's' digunakan untuk menampung hasil operasi sinus dari data set yang di tampung oleh variable x

```python
s = np.sin(2 * np.pi * x)
s
```

-Selanjutnya kita akan melakukan flotting dari kedua data tersebut 

-Pada sumbu x akan mempresentasikan nilai dari variabel x

-dan pada sumbu y akan mempresentasikan nilai dari sinus

-Kita akan menggunakan fungsi OO Style

-Tabel pertama akan menampung Figure dan tabel kedua akan menampung axes 

```python
fig, ax = plt.subplots()

ax.plot(x, s)
ax.set(xlabel ='nilai x',
       ylabel ='nilai sin(x)',
       title = 'Visualisasi nilai sin')
ax.grid()

plt.show()
```

## Multiple Subplots

-Menyiapkan 4 buat deret bilangan yang di tampung didalam 4 variabel yaitu x1, x2, y1, y2

-nilai variable x1 dan x2 akan menggunakan fungsi np.linspace
untuk nilai maks dan min nya akan sama sama bersifat inklusif

-nilai variabel y1 dan y2 akan menggunakan fungsi np.cos

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

## Multiple Subplots dengan OO Style

-Dalam 1 Figure terdapat 2 buah axes 

-Menyertakan 2 buah parameter (2, 1) yang pertama sebagai baris yang kedua sebagai kolom

-'ro-' sebagai warna yang ingin kita gunakan yaitu red (merah)

-'g.-' sebagai warna hijau

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai x')

plt.show()
```

## Multiple Subplots dengan pyplot Style

-Menggunakan plt.subplot

-Menyiapkan 3 buah Variable, 2 parameter pertama untuk mengatur layout dan parameter ketiga untuk mengendalikan sebuah axes

-'ro-' sebagai warna yang ingin kita gunakan yaitu red (merah)

-'g.-' sebagai warna hijau

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel ('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.title('Nilai $cos(x_2)$')
plt.ylabel ('Nilai $x$')

plt.show()
```
