DUA CARA DALAM MENGGUNAKAN MATPLOTLIB

2. PYPLOT STYLE

```python
x = np.linspace(0, 4, 100) # akan terjadi output sebanyak 100 buah yang di mulai dari 0 sampai 4.
x
#ini contoh yang oo Style
```

```python
fig, ax = plt.subplots()

#Contoh membuat 3 varian data axes
ax.plot(x, x, label='linear')
ax.plot(x, x**3, label='cubic')
ax.plot(x, x**5, label='quadratic')

ax.set_xlabel('x label') #Melibatkan a-label di dalam axes
ax.set_ylabel('y label') #Melibatkan y-label di dalam axes
ax.set_title("Simpel Plot") #Melibatkan title dalam axes
ax.legend() #Melibatkan legend
```

ini adalah hasil outputnya :
keterangan : 
linear (biru)

cubic (orange)

quadratic (Hijau)

tanda operator** tadi di artikan perpangkatan.
