## Sample dataset

- menggunakan fungsi np.linspace

- np.meshgrid digunakan untuk membentuk sebuah permukaan atau area x dan y

- z = np.sqrt digunakan untuk menampung nilai akar kuadrat

```python
n = np.linspace(-3.0, 3.0, 100)


x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## Contour plot 

- merupakan visualisasi data dari 3d menjadi 2d

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

## FIlled Contour Plot

- memberi warna pada graph yang di hasilkan

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
