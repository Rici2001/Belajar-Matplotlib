## Sample data

- mempersiapkan data yang dibentuk dari dictionary

- struktur dictionary merupakan struktur data yang terdiri dari pasangan keys dan values

```python
data = {'Item A': 109438,
        'Item B': 103569,
        'Item C': 112214,
        'Item D': 112591,
        'Item E': 100934,
        'Item F': 103660,
        'Item G': 137351,
        'Item H': 123381,
        'Item I': 135841,
        'Item J': 104437}

data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

## Simple plot

- menggunakan fungsi plt.subplots

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

## Pengaturan Style

- plt.style.available digunakan untuk mengecek style apa aja yang bisa digunakan

```python
plt.style.available
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

## Pengaturan Tick Label 

- parameter alignment diberi nilai right

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set property
plt.setp(labels,
         rotation=45,
         horizontalalignment='right')

plt.show()
```

## Pengaturan Format pada Ticker



```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot',
       xlabel ='Jumlah',
       ylabel ='Items')

plt.show()
```

## Pengaturan Label pada Sumbu axis dan judul title

- dengan memanggil method save

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot',
       xlabel ='Jumlah',
       ylabel ='Items')

```

## Penambahkan Garis vertikal /horizontal pada plot

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(80000, 
         ls = '-',
         color ='green')

ax.set(title='Contoh Plot',
       xlabel ='Jumlah',
       ylabel ='Items')

ax.set(title='Contoh Plot',
       xlabel ='Jumlah',
       ylabel ='items')

plt.show()
```

## Menyimpan Hasil Plot ke dalam suatu File

- mengecek format 

- format yang di gunakan png

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
         ls = '--',
         color ='r')

ax.set(title='Contoh Plot',
       xlabel ='Jumlah',
       ylabel ='Items')

fig.savefig('Sales.png', 
            transparent=False,
            dpi = 8)
plt.show()
```
