## Sample Dataset

- random digunakan untuk menampilkan sejumlah nilai yang teracak

- loc = nilai tengah

- scale = nilai standart deviasi

- size = banyak dataset yang ingin di buat

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

## Simple Box Plot

- plt.title berfungsi untuk menentukan nama tabel 

```python
plt.boxplot(data, labels=['Data'])

plt.title('Sample Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Notched Box Plot

- plt.title berfungsi untuk menentukan nama tabel

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Sample Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Pengaturan marker untuk outlier

- showfliers apa bila bernilai True akan memunculkan outliernya begitu juga sebaliknya

- markerfacecolor digunakan untuk menentukan warna yang diinginkan

```python
plt.boxplot(data,
            labels=['Data'],
            showfliers=True,
            flierprops={'markerfacecolor':'r', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Horisontal Box Plot

- menggunakan parameter vert = vertikal yang bernilai False

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Multiple Box Plot

- mempersiapkan 4 buah variabel 

- setiap variabel menampung nilai yang akan di distribusi

- variabel labels digunakan untuk menampung data String

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data1', 'Data2', 'Data3', 'Data4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
