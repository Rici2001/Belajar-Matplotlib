## Membuat plotting sederhana

-Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan satu atau lebih deret bilangan (list/array)

-Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan itu akan di jadikan sebagai data point untuk sumbu Y, Sedangkan sumbu X akan terbentuk secara otomatis dengan nilai yang berurutan dari angka 0

-Pemanggilan plt.show() berfungsi agar hasil output kita lebih bagus dilihat, teman teman semua bisa mencobanya apabila tidak menggunakan plt.show()

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

-Kita kanan mencoba merealisasikan nilai yang melewatkan 2 buah deret bilangan.

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

## Pengaturan Format pada plot

-Selain menyertakan dua buah deret bilangan sebagai parameter sumbu X dan Y, terdapat juga parameter ketiga yang digunakan untuk mengatur warna dan jenis plotting

-Pengaturan ketiga ini sepenuhnya menggunakan gaya formatting pada MATLAB

-Nilai default untuk parameter ketiga ini adalah 'b-'

-Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax]

-Parameter pertama [1, 2, 3, 4] akan kita berkorelisasi dengan sumbu X

-Parameter kedua [1, 4, 9, 16] akan kita berkorelisasi dengan sumbu Y

-Dan parameter ketiga 'ro' menunjukan warna merah. itu bukan 0 tetapi o, dan huruf o bisa di ganti menjadi * dan output yang dihasilkan pun akan ikut berubah menjadi bintang

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

## Multi plot dalam satu pemanggilan fungsi plot()

-Kali ini kita akan menggunakan np.arange dan kita membutuhkan 3 parameter (0., 5., 0,2)
0 sebagai parameter pertama, 5 sebagai parameter kedua dan 0,2 sebagai perameter ketiga

-di dalam arange parameter ketiga ini berguna untuk mengatur interval nilai

```python
t = np.arange(0., 5., 0.2) 
t
```

-Kali ini kita akan memanggil plt.plot sebanyak 3 kali artinya kita akan menjadi 3 jenis plot.

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot()

-menggunakan tanda koma (,) untuk melanjutkan ke plotting berikutnya.

-Dan hasil output yang dikeluarkan akan sama persis dengan hasil yang sebelumnya.

```python
plt.plot(t, t, 'r--', 
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

## Plotting dengan keywords

-Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan Pandas DataFrame

-String a akan berkorelisasi dengan np.arange yang berjumlah 50

-String c akan berkorelisasi dengan np.random.randInt yang akan menampilankan output nilai dengan posisi teracak

-String d akan berkorelisasi dengan np.random.randn akan menampilkan nilai yang teracak dengan jangkauan 50

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) *100

data
```

_menggunakan Scatter plot:

```python
plt.scatter('a', 'b', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

_nilai yang ditampung di 'd' akan di dipresentasikan sebagai ukuran dan semakin besar nilai d nya maka ukuran dari data point akan semakin besar.

-nilai yang ditampung di 'c' akan dipresentasikan sebagai warna 

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## Plotting untuk tipe data categorical

-Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data atau categorical variable

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

_Menggunakan cara scatter

```python
plt.scatter(names, values)
plt.show()
```

_Menggunakan cara barplot 

```python
plt.bar(names, values)
plt.show()
```
